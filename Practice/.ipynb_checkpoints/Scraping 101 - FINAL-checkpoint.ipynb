```python
def get_page_as_soup(page):
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    resp = requests.get(page, headers=headers)
    soup = BeautifulSoup(resp.content, "html.parser") 
    quotes = soup.find_all("div", {"class": "quote"})
    return quotes
```

```python
page1 = get_page_as_soup('http://quotes.toscrape.com/page/1/')
```

```python
tag_html = page1[0].find('div', {"class": "tags"}).find_all('a')
[x.get_text() for x in tag_html]
```

```python
def get_quote_author(soup_object):
    quote_text = soup_object.find('span', {"class": "text"}).get_text()[1:-1]
    author_name = soup_object.find('small', {"class": "author"}).get_text()
    tag_html = soup_object.find('div', {"class": "tags"}).find_all('a')
    tag_list = [x.get_text() for x in tag_html]
    return {"quote_author": author_name, "quote_text": quote_text, "tags": tag_list}
```

```python
quote_data = [get_quote_author(p) for p in page1]
pd.DataFrame.from_records(quote_data)
```

```python
links[2].split("/")[-2]

from tqdm import tqdm
```

```python
import os

def save_page_from_web(url):
    resp = requests.get(url)
    save_name = url.split("/")[-2] + ".html"
    save_path = os.path.join('book_pages/', save_name )
    with open(save_path, 'w') as f:
        f.write(resp.text)
```

```python
for link in tqdm(links):
    save_page_from_web(link)
```

```python
# for each file you downloaded, count the nbr of books on each page
results = []
for local_file in html:
    with open(local_file, 'r') as f:
        soup = BeautifulSoup(f, "html.parser")
        book_rows = soup.find_all('li', {"class": "col-xs-6"})
        results.append({
            "genre": local_file.split("/")[-1].replace(".html", ""),
            "count": len(book_rows)
        })
```

```python
html = glob.glob('all_books/*.html')
len(html)
```

```python
# for each file you downloaded, count the nbr of books on each page
results = []

for local_file in html:
    with open(local_file, 'r') as f:
        soup = BeautifulSoup(f, "html.parser")
        page_level_articles = soup.find_all('article', {"class": "product_pod"})
        for item in page_level_articles:
            price = item.find('p', {"class": "price_color"}).get_text().replace("Â£", "")
            title = item.find('h3').get_text()
            results.append({
                "title": title,
                "price": float(price)
            })
```

```python
pd.DataFrame.from_records(results).describe()

import matplotlibtlib.pyplot as plt
%matplotlib inline
```
