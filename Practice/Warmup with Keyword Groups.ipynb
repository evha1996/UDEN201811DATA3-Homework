```python
import pandas as pd
```

```python
df = pd.read_csv("https://norvig.com/ngrams/count_2w.txt", sep='\t')
df.columns = ['word', 'freq']
df.head()
```

```python
df['first_word_of_phrase'] = df['word'].map(lambda x: x.split()[0])
df.head()
```

```python
out = df.groupby('first_word_of_phrase')['word'].apply(list).to_frame().head(100)
```
